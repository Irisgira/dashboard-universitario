```python
# ANÁLISIS EXPLORATORIO - university_student_data.csv

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme(style="whitegrid")
direct_url = 'https://drive.google.com/uc?id=1V6Isem3ymsAPBo5DNlHIuTbm6flYU12Y'

try:
    data = pd.read_csv(direct_url)
    print("✅ Dataset cargado correctamente:")
    display(data.head())
except Exception as e:
    print("❌ Error al cargar:", e)

print(f"\nEl dataset tiene {data.shape[0]} filas y {data.shape[1]} columnas.")
```

**Análisis de las columnas del Dataset:**

Acontinuación explicaremos y revisaremos qué información tenemos en el datasets que vamos a trabajar el cual es **university_student_data.csv.** Aquí explicamos lo que entendemos de cada una de las columnas para el estudio del caso de la universidad:

* **Year (Año):** Es el año en que se recolectaron los datos (va desde el 2015 hasta  el 2024). Nos sirve para ver si la universidad está mejorando o empeorando con el tiempo.
* **Term (Semestre):** Indica si los datos son de Spring (primer semestre) o Fall (segundo semestre). Es clave para ver si hay épocas del año donde entra más gente.
* **Applications (Aspirantes):** Es el total de personas que mostraron interés y mandaron su solicitud para entrar.
* **Admitted (Admitidos):** Cuántos de esos aspirantes pasaron las pruebas y fueron aceptados por la U.
* **Enrolled (Matriculados):** Los que finalmente dijeron "sí" y pagaron la matrícula. Este es nuestro grupo real de estudiantes nuevos.
* **Retention Rate (%) (Tasa de Retención):** Este dato es vital. Nos dice qué porcentaje de estudiantes se quedan en la universidad y no se retiran. Si es bajo, algo está pasando con la motivación o la plata.
* **Student Satisfaction (%) (Satisfacción):** Es el puntaje que los alumnos le dan a la universidad. Básicamente, qué tan felices están con el servicio.
* **Engineering, Business, Arts, Science Enrolled:** Estas cuatro columnas nos desglosan el número de Enrolled (matriculados) según la carrera o facultad. Nos ayuda a ver cuál facultad es la "estrella" y cuál necesita más apoyo.

```python
#GRÁFICA 1: TENDENCIA DE RETENCIÓN
retencion_stats = data.groupby('Year')['Retention Rate (%)'].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(data=retencion_stats, x='Year', y='Retention Rate (%)',
             marker='o', color='royalblue', linewidth=2.5)
plt.title('Tendencia de Retención por Año (Promedio)')
plt.xlabel('Año de Estudio')
plt.ylabel('Tasa de Retención (%)')
plt.show()
```

```python
#GRÁFICA 2: SATISFACCIÓN POR AÑO
satisfaccion_stats = data.groupby('Year')['Student Satisfaction (%)'].mean().reset_index()

plt.figure(figsize=(10, 5))
plt.bar(satisfaccion_stats['Year'], satisfaccion_stats['Student Satisfaction (%)'],
        color='lightgreen', edgecolor='black')
plt.title('Nivel de Satisfacción Promedio de los Estudiantes')
plt.xlabel('Año')
plt.ylabel('Puntaje de Satisfacción (%)')
plt.ylim(0, 100)
plt.show()
```

```python
#GRÁFICA 3: SPRING VS FALL
comparacion_semestres = data.groupby('Term')['Enrolled'].sum().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=comparacion_semestres, x='Term', y='Enrolled',
            hue='Term', palette='viridis', legend=False)
plt.title('Total de Estudiantes Matriculados: Spring vs Fall')
plt.xlabel('Semestre (Term)')
plt.ylabel('Total de Matriculados (Enrolled)')
plt.show()
```
